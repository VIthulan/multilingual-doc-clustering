```python
import pandas as pd
import json
import glob
```

```python
files = glob.glob("sample_data/*.json")
```

```python
data = []
for file in files:
    with open(file) as f:
        doc = json.load(f)
        doc['file_name'] = file
        data.append(doc)
```

```python
df = pd.DataFrame(data)
```
